# Estadística inferencial

Las estadísticas inferenciales generalmente implican generar deducciones y / o predicciones sobre una población. 

En varios casos, se hacen inferencias sobre una población utilizando una muestra. 

A diferencia de la estadística descriptiva donde se describe una muestra conocida / datos de población, la estadística inferencial utiliza datos de muestra para sacar conclusiones sobre la población.
