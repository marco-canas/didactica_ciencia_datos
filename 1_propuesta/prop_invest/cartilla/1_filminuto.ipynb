# Metodología para filminuto

La secuencia de enseñanza que proponemos para la enseñanza del machine learning es: 

1. Python para ciencia de datos. 
2. elementos de estadística para ciencia de datos.
3. Enseñar a construir modelos de regresión  
4. Enseñar a construir modelo de clasificación. 

Este modelo educativo utiliza un enfoque de aula invertida. 


# Tips

1. Fomentar la interacción y colaboración vinculando las videoclases a los cuadernos Jupyter. 

2. Proporcionar retroalimentación efectiva con Google Colab. 

3. Estimular la autonomía y la experimentación enseñando el lenguaje Python.
