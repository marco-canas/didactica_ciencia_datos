# Metodología para filminuto

Para la enseñanza del Machine Learning seguimos una secuencia de enseñanza en la que partas de presentar : 

1. Python y estadística para ciencia de datos. 
2. Enseñar a construir modelos de regresión  
3. Enseñar a construir modelo de clasificación. 

Este modelo educativo utiliza un enfoque pedagógico de aula invertida. 


# Tips

Sugerimos para la implementación de la enseñanza del machine learning

1. Fomentar la interacción y colaboración vinculando las videoclases a los cuadernos Jupyter. 

2. Proporcionar retroalimentación efectiva utilizando la aplicación en línea: "Google Colab". 

3. Estimular la autonomía y la experimentación enseñando el lenguaje Python.
