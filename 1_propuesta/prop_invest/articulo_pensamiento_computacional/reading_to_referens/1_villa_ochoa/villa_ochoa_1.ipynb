El estudio de caso es una metodología de investigación cualitativa que se utiliza para examinar detalladamente un fenómeno particular, un evento o una situación en su contexto real. En el caso de investigar la manera de integrar el pensamiento computacional en el programa de Administración de Empresas, el estudio de caso permitiría analizar de forma profunda y holística cómo esta integración se lleva a cabo en una institución educativa o en un grupo específico de estudiantes y profesores.



A continuación, se describe cómo se lleva a cabo un estudio de caso en la investigación sobre la integración del pensamiento computacional en el programa de Administración de Empresas:



1. **Definición del caso**: Lo primero es seleccionar el caso específico que será objeto de estudio. Puede ser una universidad en particular, un departamento dentro de esa universidad, o incluso un grupo de estudiantes y profesores involucrados en el programa de Administración de Empresas.



2. **Diseño de la investigación**: Se define el enfoque y los objetivos del estudio de caso. En este caso, el objetivo sería comprender cómo se está llevando a cabo la integración del pensamiento computacional en el programa y qué impacto está teniendo en los estudiantes y el proceso educativo.



3. **Recopilación de datos**: Se utilizan diversas técnicas de recopilación de datos para obtener información relevante. Entre ellas se incluyen entrevistas con profesores y estudiantes, observación de clases y actividades, análisis de documentos y planes de estudio, y posiblemente encuestas para recopilar datos cuantitativos.

4. **Análisis de datos**: Los datos recopilados se analizan de manera rigurosa para identificar patrones, temas y tendencias relevantes. Es importante utilizar un enfoque inductivo para permitir que emerjan nuevas ideas y conceptos a partir de los datos.

5. **Elaboración del informe**: Una vez que se ha realizado el análisis, se elabora un informe detallado que presenta los hallazgos del estudio de caso. El informe debe ser claro y estructurado, y debe incluir ejemplos y citas directas de los participantes para respaldar los resultados.

6. **Discusión y conclusiones**: Se discuten los hallazgos en relación con los objetivos del estudio y se extraen conclusiones relevantes. Además, se puede sugerir recomendaciones para mejorar la integración del pensamiento computacional en el programa de Administración de Empresas.



Un estudio de caso bien realizado puede proporcionar información valiosa sobre cómo se está implementando el pensamiento computacional en el programa de Administración de Empresas, qué enfoques son más efectivos, qué desafíos se enfrentan y cómo se puede mejorar la integración en el futuro. Es importante destacar que los resultados de un estudio de caso no son generalizables, ya que se centran en un caso particular, pero pueden servir como base para futuras investigaciones y para informar sobre prácticas educativas en otros contextos similares.

# Áreas del sistema escolar priorizadas


En cuanto a las disciplinas en las cuales se desarrollan experiencias para integrar el Pensamiento Computacional (PC), los resultados muestran que las clases de informática/tecnología (diecinueve estudios) y matemáticas (siete estudios) tienen la mayor frecuencia de implementación de estrategias relativas al PC. 

En el área de informática/tecnología, principalmente, se observaron actividades relacionadas con la creación de algoritmos computacionales en lenguajes de programación en los cuáles la escritura del código es necesaria, como Python o C++, y solo uno de los estudios integró una estrategia desenchufada a través de un juego de mesa; en matemáticas predominaron actividades de resolución de problemas donde las habilidades
del PC eran un medio para el estudio de contenidos matemáticos y, en menor medida, hubo diseños para el estudio de algoritmos y el uso de las matemáticas para resolver problemas o comprender contextos de la computación.


Por otra parte, se registraron experiencias (dos estudios) que integran el PC en las clases
de humanidades, en las cuales el diseño de videojuegos y el storytelling fueron el foco de
enseñanza para el desarrollo de conocimientos tanto de historia como de PC. Con estos
hallazgos se evidencia que, en la actualidad, la integración del PC se presenta
principalmente en las áreas relacionadas con las ciencias de la computación, a pesar de
reconocerse su importancia en otros campos disciplinares (Villa-Ochoa y Castrillón-Yepes,
2020). En este sentido, la integración del PC en asignaturas diferentes a las relacionadas
con las ciencias de la computación sigue constituyéndose como un reto en la región.

¿Cuáles son las habilidades del pensamiento computacional que se pueden desarrollar en estudiantes de Administración de empresas y Contaduría?  

El pensamiento computacional es una habilidad fundamental que se puede desarrollar en estudiantes de Administración de Empresas y Contaduría para mejorar su capacidad de resolver problemas y tomar decisiones informadas en un entorno empresarial cada vez más impulsado por la tecnología y los datos. Algunas de las habilidades del pensamiento computacional que se pueden cultivar en estos estudiantes incluyen:

1. **Resolución de problemas algorítmicos**: Los estudiantes pueden aprender a abordar problemas complejos descomponiéndolos en pasos más pequeños y definir algoritmos para resolverlos eficientemente.

2. **Pensamiento lógico y abstracción**: Desarrollar la capacidad de razonar lógicamente y abstraer los elementos clave de un problema, permitiendo identificar patrones y relaciones entre datos y conceptos.

3. **Diseño y optimización de procesos**: Los estudiantes pueden aprender a aplicar el pensamiento computacional para diseñar y optimizar procesos empresariales, identificando ineficiencias y proponiendo mejoras.

4. **Análisis de datos y toma de decisiones basada en datos**: A través del análisis de datos y herramientas de ciencia de datos, los estudiantes pueden aprender a tomar decisiones fundamentadas en información cuantitativa y cualitativa.

5. **Automatización y eficiencia**: Los estudiantes pueden aprender a utilizar la programación y herramientas computacionales para automatizar tareas repetitivas y mejorar la eficiencia operativa.

6. **Modelamiento y simulación**: El pensamiento computacional puede ayudar a los estudiantes a construir modelos y simulaciones para analizar escenarios complejos y predecir resultados en el ámbito empresarial.

7. **Colaboración y trabajo en equipo**: Los estudiantes pueden aprender a trabajar de manera colaborativa en proyectos de pensamiento computacional, fomentando el trabajo en equipo y la comunicación efectiva.

8. **Creatividad e innovación**: Al abordar problemas con herramientas de pensamiento computacional, los estudiantes pueden desarrollar su creatividad e innovación al buscar soluciones no convencionales.

9. **Comunicación y visualización de datos**: Aprender a presentar resultados de análisis de datos y modelos de manera clara y visualmente atractiva para una mejor comunicación y comprensión.

10. **Adaptabilidad tecnológica**: Desarrollar la habilidad de adaptarse rápidamente a nuevas tecnologías y herramientas para mantenerse actualizado en un mundo empresarial en constante evolución.

Estas habilidades del pensamiento computacional son valiosas en el campo de la Administración de Empresas y la Contaduría, ya que permiten a los estudiantes abordar desafíos complejos, aprovechar oportunidades de negocios y tomar decisiones estratégicas basadas en datos sólidos. Al incorporar el pensamiento computacional en su formación, los estudiantes estarán mejor preparados para enfrentar los retos del mundo empresarial y destacar en un mercado laboral altamente competitivo y tecnológicamente avanzado.
