# La librería NumPy de Python y su estructura de datos insignia que es el nbarray

```python
import numpy as np 
```

## Crear estructuras de datos con numpy

```python
estaturas_list = [1.65, 1.70, 1.75, 1.64, 1.60]
```

```python
edades_list = [48, 21, 21, 21, 24]
```

```python
len(estaturas_list) == len(edades_list) 
```

```python
estaturas_array = np.array(estaturas_list)
```

```python
edades_array = np.array(edades_list) 
```

```python
type(estaturas_list)
```

```python
type(estaturas_array)
```

# La dimensión de los array 

```python
estaturas_array.ndim
```

```python
edades_array.ndim
```

# Acceder a la información de una array

```python
estaturas_array[5]
```

```python
estaturas_array[-1] # así, con el -1 accedes al ultimo dato
```

```python
estaturas_array[-2]
```

```python
for edad in edades_array:
    print(edad)
```

```python
for i in range(5):
    print(edades_array[-i-1])
```

# crear tablas con NumPy

```python
tabla = np.array([edades_list, estaturas_list]) # se crean a través de la función np.array() evaluada en listas de python
```

```python
tabla 
```

#  Calcular la dimensión de tu arreglo

```python
tabla.ndim
```

# Acceso a los datos en una matriz o tabla dos dimensional

`tabla[<fila>,<columna>]`


```python
tabla[0,0]
```

```python
tabla[1,2]
```

# Cómo accedes a filas de una array dos dimensional o tabla

```python
tabla[0,:]
```

```python
tabla[1, :]
```

# Cómo accedes a columnas de una array dos dimensional o tabla

```python
tabla[:,0] # primera columna
```

```python
tabla
```

```python
tabla[:,1] # obtuvimos la segunda columna
```
