# Tuplas 

Las tuplas son también una estructura de datos que a diferencia de las listas, una vez creadas no pueden ser modificadas. 

## Creación de tuplas

```python
tupla_datos_de_mi_hijo = (nombre_de_mi_hijo, edad_de_mi_hijo, estatura_de_mi_hijo, es_mayor_que(47, 14))
```

```python
tupla_datos_de_mi_hijo
```

## Cómo acceder a los elementos de una tupla

```python
tupla_datos_de_mi_hijo[3]
```

## Cómo hacer una copia de una tupla

```python
tupla = ('Hola', 'Mundo', 'somos', 'tupla') 
```

```python
print(tupla)
```

Note que la tupla utiliza los paréntesis redondos para delimitar sus valores.  

Las tuplas a diferencia de las listas tienen bastante menos métodos asociados a ellas. 

# Uso del método `.count()`

```python
tupla.count('Hola') 
```

```python
tupla.index('Hola') 
```

Veamos que las tuplas no son modificables como las listas, pues por ejemplo no tienen asociado el método `.append()`

```python
tupla.append('chanchito') 
```

Para transformar una tupla, podemos transformarla en una lista utilizando la función de conversión de tipos `list()`

```python
listaDeTupla = list(tupla) 
```

```python
listaDeTupla.append('chanchito')
```

```python
tupla = tuple(listaDeTupla) 
```

```python
tupla 
```
