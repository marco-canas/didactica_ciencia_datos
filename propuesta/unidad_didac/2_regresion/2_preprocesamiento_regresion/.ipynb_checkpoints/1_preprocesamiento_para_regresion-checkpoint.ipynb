# Modelo general de clase: [aula invertida](https://www3.gobiernodecanarias.org/medusa/ecoescuela/pedagotic/aprendizaje-invertido-flipped-classroom/)

## Definición y características de este modelo

### ¿Qué es?  

Flipped Learning o aprendizaje invertido es un enfoque pedagógico en el que:  

* la instrucción directa se realiza fuera del aula y 
* se utiliza el tiempo de clase para llevar a cabo actividades que impliquen el desarrollo de procesos cognitivos de mayor complejidad, en las que son necesarias la ayuda y la experiencia del docente.

## ¿Para qué sirve?  

Mejora el proceso de enseñanza-aprendizaje realizando,   

* fuera del aula, actividades de aprendizaje sencillas (observar, memorizar, resumir, etc.) y, 
* en el aula, actividades más complejas (razonar, examinar, priorizar, argumentar, proponer, etc.) que requieren la interacción entre iguales y la ayuda del docente como facilitador.

# Proceso Flipped Learning

## Fuera del Aula

### Docente:

* Diseña y planea el proceso de enseñanza-aprendizaje del aula, integrando estrategias, modelos de enseñanza y metodologías (aprendizaje basado en proyectos, aprendizaje cooperativo y colaborativo,…).  

* Diseña y planea actividades que se realizarán fuera del aula (ver vídeos, cumplimentar cuestionarios,…).  

* Selecciona e integra la tecnología en actividades de aprendizaje activo.  

* Diseña actividades de evaluación que promuevan el aprendizaje del alumnado.

### Alumnado:

* Accede, cuantas veces necesite, a las actividades facilitadas por el profesorado.
* Realiza las actividades propuestas.

## En el Aula

### Docente:

* Guía y facilita los procesos de aprendizaje, atendiendo a la diversidad.
* Retroalimenta el desempeño de los equipos/grupos y del alumnado individualmente.
* Genera espacios para la coevaluación y autoevaluación.

### Alumnado:

* Realiza las actividades (debates, exposiciones orales, informes, diario de aprendizaje, cuaderno de trabajo cooperativo,…).
* Recibe retroalimentación inmediata del profesorado y de sus compañeros y compañeras.
* Accede, si fuera necesario, a las actividades previas facilitadas por el profesorado.

2. ¿Qué es regresión?  

Las tareas de regresión se refieren a la predicción de valores numéricos continuos en función de variables de entrada. Aquí tienes algunos ejemplos de tareas de regresión:

    1. Predicción del precio de una vivienda en función de características como el número de habitaciones, el tamaño del terreno, la ubicación, etc.

    2. Estimación del consumo de energía de un edificio en función de variables como el tamaño, el número de ocupantes, la ubicación geográfica, etc.

    3. Pronóstico de la demanda de productos en función de variables como el histórico de ventas, el precio, las promociones, etc.

    4. Predicción del rendimiento de un estudiante en función de variables como el tiempo de estudio, las calificaciones anteriores, el nivel de motivación, etc.

    5. Estimación de la edad de una persona en función de variables biométricas como la altura, el peso, las medidas corporales, etc.

    6. Predicción de la velocidad del viento en función de variables atmosféricas como la temperatura, la humedad, la presión, etc.

    7. Estimación del tiempo de entrega de un paquete en función de variables como la distancia, el tipo de transporte, las condiciones climáticas, etc.

    8. Pronóstico del precio de una acción en el mercado de valores en función de variables como los precios históricos, los volúmenes de negociación, los eventos económicos, etc.

    9. Predicción de la duración de un proyecto en función de variables como la complejidad, los recursos asignados, el número de personas involucradas, etc.

    10. Estimación del valor de reventa de un automóvil en función de variables como la marca, el modelo, el año de fabricación, el kilometraje, etc.

Estos son solo algunos ejemplos de tareas de regresión. En general, cualquier problema en el que se busque predecir o estimar un valor numérico continuo se puede abordar mediante técnicas de regresión.

3. ¿Qué es una variable categórica?  
Las variables categóricas son aquellas que representan una característica o atributo que puede pertenecer a un conjunto finito y discreto de categorías o grupos. Aquí tienes algunos ejemplos de variables categóricas:

    1. Género: Puede tener dos categorías, "masculino" y "femenino".

    2. Color favorito: Puede tener múltiples categorías como "rojo", "azul", "verde", etc.

    3. Estado civil: Puede tener categorías como "soltero", "casado", "divorciado" o "viudo".

    4. Tipo de vivienda: Puede tener categorías como "casa", "apartamento", "dúplex", etc.

    5. Nivel educativo: Puede tener categorías como "primaria", "secundaria", "licenciatura", "maestría", "doctorado", etc.

    6. Nacionalidad: Puede tener categorías como "estadounidense", "francés", "japonés", etc.

    7. Orientación política: Puede tener categorías como "conservador", "liberal", "socialista", "anarquista", etc.

    8. Tipo de transporte utilizado: Puede tener categorías como "automóvil", "bicicleta", "autobús", "metro", etc.

    9. Nivel de satisfacción: Puede tener categorías como "muy satisfecho", "satisfecho", "neutral", "insatisfecho", "muy insatisfecho".

    10. Clasificación de productos: Puede tener categorías como "electrónica", "ropa", "alimentos", "muebles", etc.

Estos son solo algunos ejemplos de variables categóricas. En general, cualquier atributo o característica que pueda ser clasificada en categorías discretas puede considerarse una variable categórica.

## Download the Data

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()
```

## Create a Test Set

```python
import numpy as np

def shuffle_and_split_data(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = shuffle_and_split_data(housing, 0.2)
len(train_set)
```
