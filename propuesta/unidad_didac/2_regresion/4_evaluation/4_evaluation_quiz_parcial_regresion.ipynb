## Definición y características de este modelo

## ¿Qué es?  

Flipped Learning o aprendizaje invertido es un enfoque pedagógico en el que:  

* la instrucción directa se realiza fuera del aula y 
* se utiliza el tiempo de clase para llevar a cabo actividades que impliquen el desarrollo de procesos cognitivos de mayor complejidad, en las que son necesarias la ayuda y la experiencia del docente.

## ¿Para qué sirve?  

Mejora el proceso de enseñanza-aprendizaje realizando,   

* fuera del aula, actividades de aprendizaje sencillas (observar, memorizar, resumir, etc.) y, 
* en el aula, actividades más complejas (razonar, examinar, priorizar, argumentar, proponer, etc.) que requieren la interacción entre iguales y la ayuda del docente como facilitador.

# Proceso Flipped Learning

## Fuera del Aula

### Docente:

* Diseña y planea el proceso de enseñanza-aprendizaje del aula, integrando estrategias, modelos de enseñanza y metodologías (aprendizaje basado en proyectos, aprendizaje cooperativo y colaborativo,…).  

* Diseña y planea actividades que se realizarán fuera del aula (ver vídeos, cumplimentar cuestionarios,…).  

* Selecciona e integra la tecnología en actividades de aprendizaje activo.  

* Diseña actividades de evaluación que promuevan el aprendizaje del alumnado.

### Alumnado:

* Accede, cuantas veces necesite, a las actividades facilitadas por el profesorado.
* Realiza las actividades propuestas.

## En el Aula

### Docente:

* Guía y facilita los procesos de aprendizaje, atendiendo a la diversidad.
* Retroalimenta el desempeño de los equipos/grupos y del alumnado individualmente.
* Genera espacios para la coevaluación y autoevaluación.

### Alumnado:

* Realiza las actividades (debates, exposiciones orales, informes, diario de aprendizaje, cuaderno de trabajo cooperativo,…).
* Recibe retroalimentación inmediata del profesorado y de sus compañeros y compañeras.
* Accede, si fuera necesario, a las actividades previas facilitadas por el profesorado.

Una variable adecuacuada para regresión es la temperatura puede ser tratada como una variable aleatoria continua, ya que puede tomar cualquier valor dentro de un rango de temperaturas posibles y no solo valores discretos. Además, la temperatura también puede ser modelada como una variable aleatoria normal, ya que en muchos casos sigue una distribución normal o gaussiana.

Y recuerde que variables categóricas son las apropiadas para tareas de clasificación no de regresión: En estadística, una variable categórica es una variable que toma valores de un conjunto limitado y predefinido de categorías o grupos. Estas categorías pueden ser cualitativas o cuantitativas, pero no tienen un orden inherente entre ellas.

Por ejemplo, una variable categórica podría ser el género de una persona, con dos posibles categorías: masculino o femenino. Otra variable categórica podría ser el tipo de vehículo, con varias categorías como automóvil, camión, motocicleta, etc.

Las variables categóricas son distintas de las variables numéricas, que tienen un valor numérico continuo o discreto. En el análisis estadístico, las variables categóricas se pueden analizar utilizando técnicas como tablas de contingencia, pruebas de chi-cuadrado y análisis de varianza (ANOVA).

El precio de los carros se puede tratar como una variable aleatoria continua, ya que puede tomar cualquier valor dentro de un rango de precios posibles y no solo valores discretos.

Además, el precio de los carros puede ser modelado como una variable aleatoria no negativa, ya que los precios no pueden ser negativos. Dependiendo de la distribución de precios en una población de carros, es posible que el precio pueda ser modelado utilizando una distribución normal, lognormal u otra distribución continua. En el análisis estadístico, se pueden utilizar diversas técnicas para analizar y comparar los precios de los carros, como pruebas t, ANOVA, regresión lineal, etc.
