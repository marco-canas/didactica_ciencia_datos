```python
plt.scatter(X[:,2][y==0], X[:,3][y==0], label = 'setosa')

plt.scatter(X[:,2][y!=0], X[:,3][y!=0], label = "No setosa")

plt.title("Clasificación binaria, Detector de setosa")
plt.xlabel("petal length")
plt.ylabel("petal width") 
plt.grid()

plt.legend()
```

```python
from sklearn.linear_model import Perceptron
```

```python
detector_setosa = Perceptron(random_state=16) 
```

```python
detector_setosa.fit(X[:,(2,3)], y) # aquí ya aprendió a clasificar o detectar setosa 
```

# Verificar que si clasifica o detecta setosa como queremos 

```python
if detector_setosa.predict([[4.7, 1.4]])==0:
    print('estas dimensiones de pétalo corresponden a una flor setosa')
else: 
    print('estas dimensiones de pétalo NO corresponden a una flor setosa')        
```

```python
b = detector_setosa.intercept_[0]
```

```python
m = detector_setosa.coef_
```

```python
m = m[0, 0]
m 
```

```python
b, m 
```

```python
import numpy as np 
dominio = np.array([min(X[:,2]), max(X[:,2])])
rango = m* dominio + b # y = mx + b 


plt.plot(dominio, rango) 

plt.scatter(X[:,2][y==0], X[:,3][y==0], label = 'setosa')

plt.scatter(X[:,2][y!=0], X[:,3][y!=0], label = "No setosa")

plt.title("Clasificación binaria, Detector de setosa")
plt.xlabel("petal length")
plt.ylabel("petal width") 
plt.grid()

plt.legend()

plt.savefig("detector_setosa.pdf")

plt.savefig("detector_setosa.jpg")

plt.savefig("detector_setosa.png")
```
