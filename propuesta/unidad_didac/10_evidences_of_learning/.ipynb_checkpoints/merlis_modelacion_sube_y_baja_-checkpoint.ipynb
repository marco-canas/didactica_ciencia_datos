# Modelación del Juego de Sube y Baja

## Ley de la palanca #

##planteamiento del problema ##
En la figura se ilustra un sistema de
palancas, similar al sube y baja que usted encuentra en los
parques para niños. Para que el sistema se equilibre, el producto
del peso por la distancia a partir del punto de apoyo
debe ser igual en cada lado. Es decir,
 
$$w_1x_1=w_2x_2$$

Esta ecuación recibe el nombre de ley de la palanca, y fue
descubierta por Arquímedes (véase la pág. 748).
Una mujer y su hijo están jugando en un sube y baja. El
muchacho está en un extremo, a 8 pies del punto de apoyo.
Si el hijo pesa 100 libras y la madre pesa 125 libras,  

 ¿dónde debe colocarse la mujer para equilibrar el sube y baja?


image.png



lo primero que haremos es convertir las nidades de medida peso 

**Recopilación de datos**


peso del hijo=100 libras 

peso de la madre=125 libras 

distancia a la que se encuentra el niño al punto de equilibrio =  8 pies 


**Definicion de variables** 


$M_h=$  masa en libras del hijo


$M_m=$masa de la madre 


**Longitud** 

$x_1=$Distancia a la que se encuentra el niño al punto de equilibrio.

$x_2=$ Distancia a la que se encuentra la madre del punto de equilibrio


  






##solución del problema##

Empecemos con la construcción y representación algebraica y gráfica del modelo. Y utilicemos estas representaciones para resolver o responder a la pregunta. 

Las variables a relacionar son el peso de las personas que juegan en el sube y baja y la distancia a la que se sientan con respecto al vertice del sube y baja. Así, las variables a relacionar son $w$ y $d$. 

|peso $w$|distancia $d$|  
|--|--|  
|||  

Note que las tablas markdown admiten lenguaje latex. 

```python
Mh=100
Mk = (1/2)*Mh
print (Mk)
```

Con la masa en kilogramos podemos proceder a calcular el peso del hijo(en Newtons).

```python
from sympy import *
g = 9.8 
Wh = Mk * g
print(Wh)
```

$Wm$=peso de la madre 

Convertimo las libras (125 )que pesa la madre a Kg




```python
Mm=125
Mk = (1/2)*Mm
print (Mk)
```

Con la masa en kilogramos podemos proceder a calcular el peso de la madre (en Newtons).



$Wm$

```python
from sympy import *
g = 9.8 
Wm = Mk * g
print(Wm)
```

Ahora como tenemos la distancia a la que se encuentra el niño al punto de equilibrio en pies la convertiremos a $m$ ($x_1$)

```python
p=0.35
X1=8 * p
print(X1)
```

ya que tenemos todos los datos convertidos procederemos a despejar nuestra incognita que en este caso seria $x_2$ de la siguiente ecuación

$$w_1x_1=w_2x_2$$

* Donde $w_1$ y $w_2$ son respectivamente los pesos del hijo y de la madre

* Donde $x_1$ y $x_2$ son las longitudes a las que se encuentran el hijo y la madre respectivamente .




##Reemplazamos##

para hallar el valor de $x_2$ que seria la distancia de la madre con respecto al punto de equilibrio multiplicamos ($wh$) que es el peso del niño con la distancia que es  $x1$ con respecto al punto de equilibrio y esto lo dividimos con con el peso de la madre ($Wm$).

```python
X2=Wh*X1 /Wm
print(X2)
```

#Respuesta 

la distancia a la que debe estar la madre es a 2.24 metros de longitud.

##CONCLUSIÓN 


Cuando los pesos son diferentes con respecto a un punto de equilibrio el peso mayor debe mantenerse mas cerca de este(punto de equilibrio) para que no haya un desequilibrio y la distancia del peso menor debe ser mayor.
