## Práctica de cálculo e interpretación del rango con Python

```python
import numpy as np 
import scipy as sc
import statistics as st 
```

```python
def calcula_rango_por_columnas(datos):
    maximo = np.max(datos, axis = 0)  #con axis = 0 toma los maximos por columna. 
    minimo = np.min(datos, axis = 0)
    rango_por_columnas = maximo-minimo 
    return rango_por_columnas 
```

```python
calcula_rango_por_columnas(datos) 
```

# Cálculo del rango con el la función iqr() del módulo stats de la libretría scipy

Compute the interquartile range of the data along the specified axis.

The interquartile range (IQR) is the difference between the 75th and 25th percentile of the data. It is a measure of the dispersion similar to standard deviation or variance, but is much more robust against outliers [2].

The rng parameter allows this function to compute other percentile ranges than the actual IQR. For example, setting rng=(0, 100) is equivalent to numpy.ptp.

The IQR of an empty array is np.nan.

```python
from scipy import stats
```

```python
stats.iqr(datos, axis = 1, rng=(0,100) ) # con axis = 0 calcula el rango por columnas
```

## Ejemplo de calculo de rango con uno datos de estaturas

```python
estaturas = np.array([1.65, 1.55, 1.70, 1.60, 1.65])
```

```python
estaturas.sort() 
```

```python
estaturas # estaturas ordenadas
```

```python
def rango(datos):
    maximo = np.max(datos)
    minimo = np.min(datos)
    rango = maximo-minimo
    return rango 
```

```python
rango(estaturas) 
```

```python
sc.stats.iqr(estaturas, rng=(0, 100))  
```

## Interpretación

Las alturas de las personas dadas solo difieren en 15 centímetros a lo sumo.
