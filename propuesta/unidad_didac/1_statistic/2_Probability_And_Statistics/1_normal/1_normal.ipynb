# La función  numpy.random.randn()

```python
%%time
import numpy as np 
a = np.random.randn(1000_000)
```

```python
%%time 
import matplotlib.pyplot as plt

plt.hist(a, density = True, bins = 100, histtype = 'step', color = 'blue', label = 'rand') 

plt.show() 
```

# La función np.random.randn()

```python
np.random.randn(2,4) # ojo, no le das una tupla sino dos entradas a la función numpy.random.randn()
```

```python
b = np.random.randn(100_000)
```

```python
%%time 
a = np.random.rand(10_000)
import matplotlib.pyplot as plt

plt.title('Random normal distribition')
plt.xlabel('value')
plt.ylabel('density')


plt.hist(b, density = True, bins = 100, histtype = 'step', color = 'red', label = 'randn' )
plt.savefig('distribución_normal.jpg')

plt.show() 
```

# numpy.random.normal

`random.normal(loc=0.0, scale=1.0, size=None)`

Draw random samples from a normal (Gaussian) distribution.

The probability density function of the normal distribution, first derived by De Moivre and 200 years later by both Gauss and Laplace independently, is often called the bell curve because of its characteristic shape (see the example below).

The normal distributions occurs often in nature. For example, it describes the commonly occurring distribution of samples influenced by a large number of tiny, random disturbances, each with its own unique distribution.
