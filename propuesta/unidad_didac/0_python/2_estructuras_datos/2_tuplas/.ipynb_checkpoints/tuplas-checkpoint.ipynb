## Repaso  

* Datos: los datos son información:
  - string: nombres de las personas.
  - int: edad, cantidad de estudiantes en el salon de clase. 
  - float: estatura, peso.
   
  - booleano
* Estructuras de datos
  - listas
  - tuplas

```python
nombre_de_mi_hijo = "Joseph Cañas Osorio"
```

```python
edad_de_mi_hijo = 14
```

```python
estatura_de_mi_hijo = 1.64
```

```python
def es_mayor_que(edad_1, edad_2):
    if edad_1>edad_2:
        print(True)
    elif edad_1 < edad_2:
        print(False)
    else:
        print(False)
```

```python
es_mayor_que(47,14)
```

**Conclusión:** existen cuatro tipos de datos en python: 
* cadena
* entero
* float o decimales
* booleano.
