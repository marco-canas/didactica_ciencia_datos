# if y condicionales 

* Utilizamos la palabra reservada `if`. 
* Luego le damos una proposición para que él evalue si es verdaderoa o falsa y luego termina la línea con dos puntos(:)
* Luego, en la siguiente línea indentada coiolocamos lo que queremos que haga cuendo la condición descrita en la proposición se cumpla o sea berdadera. 

```python
if 2 < 5:
    print('2 es menor que 5')
```
