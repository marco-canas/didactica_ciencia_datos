## Creación de una aplicación en python

Creemos una lista de opciones.  
y nosotros vamos a adivinar qué es lo que se encuentra dentro de esta lista.  

```python
lista = ['hola', 'mundo', 'chanchito', 'feliz', 'dragones'] 
```

Utilicemos ahora la instrucción `if`

```python
if lista.count(dato)>0:
    print('el dato existe:', dato )
else:
    print('el dato no existe', dato)
```

Aquí termina nuestro juego.
