```python
import numpy as np 
```

```python
u = np.array([2,5])
v = np.array([3,1])
```

```python
import matplotlib.pyplot as plt 

x_cords, y_cords = zip(u,v)

plt.scatter(x_cords, y_cords, color = ['r', 'b'])

plt.axis([0,5, 0,9])

plt.savefig('vectores2d_como_puntos.jpg')

plt.grid(alpha = 0.4)

plt.show()
```

# Vectores como flechas 

```python
def trazar_flechas(vector, origen = [0,0], **opcions ):
    return plt.arrow(origen[0], origen[1], vector[0], vector[1], head_length = 0.3, head_width = 0.2, length_includes_head = True,  **opcions)
```

```python
trazar_flechas(u, color = 'r')
trazar_flechas(v, color = 'b')
plt.axis([0,5, 0,9])


plt.grid(alpha = 0.4)

plt.savefig('vectores_como_flechas.jpg')

plt.show() 
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame = True) 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(iris.data.values, iris.target.values, test_size = 0.1\
                                                             ,random_state=42)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, test_size = 0.1\
                                                             ,random_state=42)
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes = [5],  random_state = 42, max_iter = 10_000 )
```

```python
pipeline = make_pipeline(StandardScaler(), mlp_clf) 
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_valid, y_valid) 
```

```python
pipeline.score(X_test, y_test)  
```

# Aprender a programar

Aprender a programar puede ser un desafío para muchas personas, pero esto no se debe a una sola razón. Algunos de los factores que pueden hacer que sea difícil aprender a programar incluyen:

* Curva de aprendizaje: Al principio, puede ser difícil comprender los conceptos y la sintaxis de un lenguaje de programación, especialmente si es la primera vez que se aprende.

* Complejidad: La programación puede ser una tarea compleja que requiere de lógica y pensamiento sistemático.
* Naturaleza abstracta: La programación es una actividad abstracta que requiere un pensamiento lógico y estructurado.
* Diferentes lenguajes: Hay muchos lenguajes de programación diferentes, cada uno con su propia sintaxis y características únicas. Esto puede ser abrumador para los principiantes.
* Falta de motivación: Aprender a programar puede ser desalentador si no se tiene un proyecto concreto en mente o si no se ve inmediatamente la aplicación práctica de los conceptos que se están aprendiendo.

Sin embargo, también es importante señalar que aprender a programar es una habilidad valiosa y que puede ser muy gratificante una vez que se comprenden los conceptos básicos. Además, existen muchos recursos disponibles, como cursos en línea, tutoriales, libros y comunidades en línea, que pueden ayudar a hacer más fácil el aprendizaje.
