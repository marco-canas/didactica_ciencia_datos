"Esta propuesta de enseñanza del machine learning con enfoque pedagógico de aprendizaje invertido se centra en la utilización del lenguaje de programación Python. Nuestro objetivo es proporcionar a los estudiantes una experiencia práctica y significativa en el campo del machine learning, combinando la teoría con la aplicación directa. 

A través de proyectos y actividades interactivas, los estudiantes aprenderán los conceptos fundamentales del machine learning y desarrollarán habilidades de programación en Python para implementar algoritmos de aprendizaje automático. Creemos que esta combinación de aprendizaje invertido y Python permitirá a los estudiantes adquirir una comprensión profunda y práctica del machine learning, preparándolos para enfrentar desafíos reales en el mundo de la inteligencia artificial."

# Abstract  

"This machine learning teaching proposal with a flipped learning pedagogical approach focuses on the use of the Python programming language. 

Our goal is to provide students with practical and meaningful experience in the field of machine learning, combining theory with application Through interactive projects and activities, students will learn fundamental machine learning concepts and develop Python programming skills to implement machine learning algorithms. 

We believe this combination of flipped learning and Python will allow students to gain a deep understanding and practice machine learning, preparing them to face real challenges in the world of artificial intelligence."

The methodological sequence followed is the python introduction, statistics for data science, regression and classification.

As a result of this didactic strategy, students:
 

1. They developed skills to obtain a dataset associated with a research or business problem.

2. They built a predictive Machine Learning model to perform a classification or regression task by estimating its performance or power of generalization. They were able to justify the way they set up their model.

3. They present the process of its construction and expose it, with the interpretations of the predictions, limitations and potentialities, which make the proposed model.

The students of the sections of the University of Antioquia throughout the department were able to access this knowledge because these learning products can be offered as courses in the regionalized programs thanks to the Flipped Learning model that configures it.

Our teaching and learning proposal is appropriate because it configures a learning environment that combines the applications: Google Colab, GitHub and YouTube that are linked to the private sector and everyday life.

# Metodología   
 

1. Introducción al lenguaje Python para ciencia de datos:
   - Presentación de los conceptos básicos del lenguaje Python, enfocados en su aplicación en la ciencia de datos.
   - Exploración de las librerías fundamentales para el análisis de datos, como NumPy, Pandas y ScikitLearn.
   - Demostración de ejemplos prácticos de manipulación y visualización de datos utilizando Python.

2. Introducción a la estadística para ciencia de datos:
   - Resumen de los conceptos estadísticos clave necesarios para el análisis de datos en machine learning.
   - Explicación de medidas descriptivas, distribuciones normal y uniforme, técnicas de muestreo.
   - Ilustración de cómo aplicar estas técnicas en Python utilizando bibliotecas como SciPy y Statsmodels.


3. Constitución de modelos de regresión para predecir variables numéricas continuas:
   - Presentación de los conceptos básicos de la regresión lineal y sus variantes.
   - Explicación de cómo entrenar y evaluar modelos de regresión utilizando técnicas como validación cruzada y métricas de evaluación.
   - Implementación de modelos de regresión en Python utilizando bibliotecas como Scikit-learn.

4. Constitución de modelos de clasificación para variables categóricas:
   - Introducción a los modelos de clasificación, los algoritmos lineales(`SGDClassifier`) y de vecinos más cercanos.
   - Implementación de modelos de clasificación en Python utilizando bibliotecas como Scikit-learn.
   - Explicación de cómo entrenar y evaluar modelos de clasificación utilizando técnicas como matriz de confusión y curvas ROC.
   

Es importante que en cada paso de la metodología se enfatice la participación activa de los estudiantes, utilizando la plataforma de enseñanza en línea asociada al correo institucional llamada Google Colab. Se debe proporcionar acceso a los contenidos del curso a través del repositorio 7_didactica_ciencia_datos en GitHub, donde los estudiantes pueden acceder a cuadernos Jupyter interactivos para la lectura experimental y práctica de los conceptos.

Además, se pueden utilizar videos asociados a los cuadernos Jupyter en YouTube para reforzar los temas tratados, brindando ejemplos prácticos y explicaciones más detalladas.

A lo largo de la ponencia, se pueden presentar ejemplos de aplicaciones reales utilizando conjuntos de datos relevantes y casos de estudio que demuestren la utilidad y aplicabilidad del machine learning en diferentes dominios.

Al seguir esta secuencia metodológica, los estudiantes podrán adquirir gradualmente los conocimientos y habilidades necesarios para comprender y aplicar el machine learning en la resolución de problemas de ciencia de datos.

# Discusión (Conclusiones) 

Hemos presentado un enfoque pedagógico innovador basado en el aprendizaje invertido, implementado a través de la enseñanza en línea utilizando la aplicación asociada al correo institucional Google Colab, el repositorio 7_didactica_ciencia_datos en GitHub y los videos asociados a los cuadernos Jupyter en en el canal de YouTube DimathData del profesor Marco Julio Cañas. A lo largo de la secuencia metodológica propuesta, que incluye la introducción al lenguaje Python para ciencia de datos, la introducción a la estadística para ciencia de datos, la constitución de modelos de regresión para predecir variables numéricas continuas y la constitución de modelos de clasificación para variables categóricas, hemos logrado alcanzar una serie de resultados de aprendizaje significativos.

El primer resultado de aprendizaje destacado es la capacidad de los estudiantes para obtener un dataset asociado a un problema investigativo o empresarial. A través de las herramientas y recursos proporcionados, los estudiantes han aprendido a comprender su estructura y características. Esta habilidad les permite abordar problemas reales y contextualizados, lo cual es fundamental en el campo del machine learning.

En segundo lugar, los estudiantes han desarrollado la capacidad de constituir modelos predictivos de machine learning para realizar tareas de clasificación o regresión, estimando su desempeño o poder de generalización. Han aprendido a aplicar técnicas de entrenamiento y evaluación de modelos, utilizando herramientas como validación cruzada y métricas de evaluación. Esta competencia les permite tomar decisiones informadas sobre la configuración de los modelos y comprender su capacidad para hacer predicciones precisas.

Además, los estudiantes han adquirido la capacidad de justificar la forma en que configuran sus modelos. Han aprendido a considerar diferentes algoritmos, ajustar parámetros y evaluar opciones de preprocesamiento de datos. Esta habilidad de justificación es esencial para garantizar la calidad y la validez de los modelos desarrollados.

Por último, los estudiantes han demostrado la capacidad de presentar el proceso de construcción de sus modelos y exponer los resultados obtenidos, incluyendo interpretaciones de las predicciones, limitaciones y potencialidades del modelo propuesto. Han aprendido a comunicar de manera efectiva sus hallazgos, tanto a través de informes escritos como de presentaciones orales, lo cual es fundamental en el ámbito de la ciencia de datos.

En resumen, a través de la implementación del enfoque pedagógico de aprendizaje invertido y el uso de herramientas tecnológicas como Google Colab, el repositorio en GitHub y los videos en YouTube, hemos logrado alcanzar importantes resultados de aprendizaje en los estudiantes. Han desarrollado habilidades para obtener datasets relevantes, constituir modelos predictivos de machine learning, justificar su configuración y presentar los resultados obtenidos. Estos logros demuestran el valor y la efectividad de este enfoque pedagógico en la enseñanza del machine learning.
