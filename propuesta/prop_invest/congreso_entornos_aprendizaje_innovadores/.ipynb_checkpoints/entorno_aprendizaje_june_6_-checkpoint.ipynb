- Uso de herramientas tecnológicas:
  - Google Colab: plataforma para la enseñanza en línea que permite la lectura experimental e interactiva.
  - GitHub: repositorio de materiales y recursos del curso en cuaderno Jupyter de extensión .ipynb.
  - YouTube: Video Clases asociados a los cuadernos Jupyter para reforzar los temas tratados.

- Metodología:
  1. Introducción a Python para ciencia de datos.
  2. Estadística para ciencia de datos.
  3. Modelos de regresión para variables numéricas continuas.
  4. Modelos de clasificación para variables categóricas.
