# [Ponencia Congreso internacional de entornos de aprendizaje innovadores](http://congresointernacional-eai.com.co/inscripciones.php)

# Título: Enseñar Machine Learning en la Universidad de Antioquia con Google Colab, GitHub y YouTube

## Resumen

* Introducción: "Esta propuesta de enseñanza del machine learning con enfoque pedagógico de aprendizaje invertido se centra en la utilización del lenguaje de programación Python. Nuestro objetivo es proporcionar a los estudiantes una experiencia práctica y significativa en el campo del machine learning, combinando la teoría con la aplicación directa. A través de proyectos y actividades interactivas, los estudiantes aprenderán los conceptos fundamentales del machine learning y desarrollarán habilidades de programación en Python para implementar algoritmos de aprendizaje automático. Creemos que esta combinación de aprendizaje invertido y Python permitirá a los estudiantes adquirir una comprensión profunda y práctica del machine learning, preparándolos para enfrentar desafíos reales en el mundo de la inteligencia artificial."



* Metodología: introducción python, estadística para ciencia de datos, regresión y clasificación.   

* Resultados:  Como resultado de esta estrategia didáctica, los estudiantes:  
 
1. Desarrollan habilidades para obtener un dataset asociado a un problema investigativo o empresarial.  

2. Constituyen un modelo predictivo de Machine Learning para realizar una tarea de clasificación o de regresión estimando su desempeño o poder de generalización.
Están en capacidad de justificar la forma en que configuran su modelo.  

3. Presentan el proceso de su construcción y lo exponen, con las interpretaciones de las predicciones, limitaciones y potencialidades, que hacen del modelo propuesto.
Los estudiantes regionales podrán acceder a estos conocimientos porque se espera que estos productos de aprendizaje se pueda ofertar como cursos en los programas regionalizados. 

* Conclusiones:  Nuestra propuesta de enseñanza y aprendizaje es apropiada pues configura un entorno de aprendizaje que conjuga las aplicaciones: Google Colab, GitHub y YouTube que están vinculadas al sector privado y a la cotidianidad.
