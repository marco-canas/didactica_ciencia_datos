## Resumen

"Esta propuesta de enseñanza del machine learning con enfoque pedagógico de aprendizaje invertido se centra en la utilización del lenguaje de programación Python. Nuestro objetivo es proporcionar a los estudiantes una experiencia práctica y significativa en el campo del machine learning, combinando la teoría con la aplicación directa. A través de proyectos y actividades interactivas, los estudiantes aprenderán los conceptos fundamentales del machine learning y desarrollarán habilidades de programación en Python para implementar algoritmos de aprendizaje automático. Creemos que esta combinación de aprendizaje invertido y Python permitirá a los estudiantes adquirir una comprensión profunda y práctica del machine learning, preparándolos para enfrentar desafíos reales en el mundo de la inteligencia artificial."

La secuencia metodológica seguida es la de introducción python, estadística para ciencia de datos, regresión y clasificación.   

Como resultado de esta estrategia didáctica, los estudiantes:  
 
1. Desarrollaron habilidades para obtener un dataset asociado a un problema investigativo o empresarial.  

2. Constituyeron un modelo predictivo de Machine Learning para realizar una tarea de clasificación o de regresión estimando su desempeño o poder de generalización. Estubieron en capacidad de justificar la forma en que configuran su modelo.  

3. Presentan el proceso de su construcción y lo exponen, con las interpretaciones de las predicciones, limitaciones y potencialidades, que hacen del modelo propuesto.  

Los estudiantes de las seccionales de la universidad de Antioquia a lo largo del departamento pudieron  acceder a estos conocimientos porque estos productos de aprendizaje se pueden ofertar como cursos en los programas regionalizados gracias al modelo de Aprendizaje invertido que lo configura. 

Nuestra propuesta de enseñanza y aprendizaje es apropiada pues configura un entorno de aprendizaje que conjuga las aplicaciones: Google Colab, GitHub y YouTube que están vinculadas al sector privado y a la cotidianidad.  

## Palabras claves:  machine learning, python, aprendizaje invertido, ciencia de datos.  

# Abstract  

"This machine learning teaching proposal with a flipped learning pedagogical approach focuses on the use of the Python programming language. Our goal is to provide students with practical and meaningful experience in the field of machine learning, combining theory with application Through interactive projects and activities, students will learn fundamental machine learning concepts and develop Python programming skills to implement machine learning algorithms. We believe this combination of flipped learning and Python will allow students to gain a deep understanding and practice machine learning, preparing them to face real challenges in the world of artificial intelligence."

The methodological sequence followed is the python introduction, statistics for data science, regression and classification.

As a result of this didactic strategy, students:
 
1. They developed skills to obtain a dataset associated with a research or business problem.

2. They built a predictive Machine Learning model to perform a classification or regression task by estimating its performance or power of generalization. They were able to justify the way they set up their model.

3. They present the process of its construction and expose it, with the interpretations of the predictions, limitations and potentialities, which make the proposed model.

The students of the sections of the University of Antioquia throughout the department were able to access this knowledge because these learning products can be offered as courses in the regionalized programs thanks to the Flipped Learning model that configures it.

Our teaching and learning proposal is appropriate because it configures a learning environment that combines the applications: Google Colab, GitHub and YouTube that are linked to the private sector and everyday life.

El aprendizaje invertido, también conocido como flipped classroom, redefine el papel tradicional del profesor y el estudiante en el proceso de enseñanza-aprendizaje. En lugar de recibir pasivamente la información en el aula, los estudiantes tienen acceso previo a los materiales de estudio, como cuadernos Jupyter interactivos, alojados en Google Colab y disponibles en nuestro repositorio en GitHub. Esto les permite explorar los conceptos y experimentar de manera autónoma antes de la clase.
