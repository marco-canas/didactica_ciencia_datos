# Resumen 

Descripción de la innovación:
La propuesta de enseñanza del machine learning en la Universidad de Antioquia se basa en un enfoque pedagógico innovador de aprendizaje invertido. Este enfoque fomenta el aprendizaje activo y autónomo de los estudiantes, quienes tienen acceso a una combinación de herramientas tecnológicas, como Jupyter de Anaconda, Google Colab, YouTube y Visual Studio Code. Estas herramientas permiten a los estudiantes explorar y aplicar conceptos de machine learning en entornos prácticos y colaborativos, facilitando la comprensión profunda de los algoritmos y técnicas de aprendizaje automático.

Proceso de implementación:
El proceso de implementación de esta propuesta implica la creación de materiales educativos interactivos y accesibles a través de plataformas como YouTube, donde los estudiantes pueden acceder a tutoriales y conferencias. Además, se utiliza Jupyter de Anaconda y Google Colab como entornos de desarrollo para prácticas y experimentación con algoritmos de machine learning. La colaboración se fomenta mediante el uso de Visual Studio Code, que permite a los estudiantes trabajar en proyectos conjuntos. Los profesores desempeñan un papel activo en la facilitación del aprendizaje y brindan orientación y retroalimentación a los estudiantes a medida que avanzan en su proceso de aprendizaje.

Evaluación de resultados:
La evaluación de los resultados se basa en criterios de rendimiento, como la comprensión teórica, la capacidad para aplicar los conceptos aprendidos en problemas reales y la colaboración en proyectos grupales. Se utilizan evaluaciones formativas y sumativas, que incluyen exámenes, proyectos prácticos y presentaciones. Además, se recopila la retroalimentación de los estudiantes para mejorar continuamente la propuesta de enseñanza y adaptarla a sus necesidades.

Conclusiones:
En conclusión, la propuesta de enseñanza del machine learning en la Universidad de Antioquia, basada en el enfoque pedagógico de aprendizaje invertido y soportada por herramientas como Jupyter de Anaconda, Google Colab, YouTube y Visual Studio Code, muestra resultados prometedores. Este enfoque proporciona a los estudiantes una experiencia de aprendizaje activa, práctica y colaborativa, lo que les permite adquirir un conocimiento profundo de los fundamentos y aplicaciones del machine learning. La combinación de recursos multimedia, entornos de desarrollo interactivos y herramientas de colaboración ha demostrado ser efectiva para fomentar la participación y el éxito de los estudiantes en esta área especializada.

The machine learning teaching proposal at the Universidad de Antioquia is based on an innovative pedagogical approach of flipped learning. This approach encourages active and autonomous learning by students, who have access to a combination of technology tools, including Anaconda's Jupyter, Google Colab, YouTube, and Visual Studio Code. These tools allow students to explore and apply machine learning concepts in hands-on, collaborative settings, facilitating a deep understanding of machine learning algorithms and techniques.

Implementation process:
The implementation process of this proposal involves the creation of interactive and accessible educational materials through platforms such as YouTube, where students can access tutorials and lectures. In addition, Jupyter from Anaconda and Google Colab are used as development environments for practices and experimentation with machine learning algorithms. Collaboration is encouraged through the use of Visual Studio Code, which allows students to work on joint projects. Teachers play an active role in facilitating learning, providing guidance and feedback to students as they progress through their learning process.

Evaluation of results:
The evaluation of the results is based on performance criteria, such as theoretical understanding, the ability to apply the concepts learned in real problems and collaboration in group projects. In addition, feedback from students is collected to continuously improve the teaching proposal and adapt it to their needs.

In conclusion, the machine learning teaching proposal at the University of Antioquia, based on the pedagogical approach of flipped learning and supported by tools such as Anaconda's Jupyter, Google Colab, YouTube and Visual Studio Code, shows promising results. This approach provides students with an active, hands-on, and collaborative learning experience, enabling them to gain a deep understanding of machine learning fundamentals and applications. The combination of multimedia resources, interactive development environments, and collaboration tools has proven effective in fostering student engagement and success in this specialized area.

3. Proceso de la implementación

Detalle las etapas seguidas para desarrollar y consolidar la innovación  

El proceso de implementación de esta propuesta de enseñanza comienza con la enseñanza del lenguaje de programación Python para ciencia de datos, basado en los fundamentos de Wes McKinney, creador de la biblioteca de Python pandas. Los estudiantes adquieren habilidades en Python, incluyendo la manipulación de datos y la aplicación de técnicas específicas para el análisis de datos.

Posteriormente, se imparte la enseñanza de la estadística para ciencia de datos, dividida en dos módulos: estadística descriptiva para ciencia de datos y estadística inferencial para ciencia de datos. En estos módulos, se exploran conceptos fundamentales de estadística, como medidas de tendencia central, dispersión y probabilidad. Además, se presenta el concepto de correlación lineal de Pearson, que es relevante en el análisis de relaciones entre variables en el contexto de la ciencia de datos.

A continuación, se procede a enseñar la construcción de modelos de regresión, siguiendo las enseñanzas de Aurelien Geron, autor reconocido en el campo del aprendizaje automático. Los estudiantes aprenden a utilizar técnicas de regresión, como la regresión lineal y la regresión logística, para predecir variables continuas como el precio de vivienda o la temperatura en Caucasia - Antioquia, respectivamente.

Finalmente, se enseña la construcción de diferentes modelos de clasificación, también de la mano de Aurelien Geron. Los estudiantes exploran y aplican algoritmos de clasificación, como árboles de decisión, máquinas de vectores de soporte (SVM) y clasificadores de vecinos más cercanos (KNN), para resolver problemas de clasificación en el campo de la ciencia de datos.

En resumen, el proceso de implementación de esta propuesta de enseñanza abarca desde la enseñanza del lenguaje de programación Python hasta la construcción de modelos de regresión y clasificación, siguiendo las enseñanzas de expertos en el campo. Los estudiantes adquieren habilidades prácticas en el manejo de datos, la estadística y la construcción de modelos, preparándolos para enfrentar desafíos reales en el ámbito de la ciencia de datos.

2. Manipulación de las principales estructuras de datos de Python: Se evalúa la capacidad de los estudiantes para manipular eficientemente las estructuras de datos en Python, como listas, tuplas, diccionarios, nd-arrays y DataFrames, utilizando las herramientas adecuadas para el análisis de datos.

3. Creación de funciones y clases de Python: Se evalúa la capacidad de los estudiantes para crear y utilizar funciones y clases en Python, con el objetivo de automatizar procesos de modelado y análisis de datos.

4. Utilización de los controladores de flujo de Python: Se evalúa la capacidad de los estudiantes para utilizar los controladores de flujo de Python, como bucles y condicionales, para el preprocesamiento de datos y la manipulación de flujos de trabajo en el análisis de datos.

5. Establecimiento de medidas de tendencia central y de variabilidad: Se evalúa la capacidad de los estudiantes para calcular y utilizar medidas de tendencia central, como la media y la mediana, así como medidas de variabilidad, como la desviación estándar, para analizar y describir las variables en un conjunto de datos.

6. Establecimiento de correlaciones lineales de Pearson: Se evalúa la capacidad de los estudiantes para calcular y analizar la correlación lineal de Pearson entre variables, utilizando técnicas estadísticas apropiadas para identificar relaciones lineales en los datos.

8. Construcción y exposición de un modelo de clasificación: Se evalúa la capacidad de los estudiantes para construir un modelo de clasificación utilizando técnicas de aprendizaje automático, y para exponer y evaluar su desempeño utilizando métricas apropiadas.

En resumen, la evaluación de resultados se basa en la presentación de evidencias de medición y valoración de los conocimientos adquiridos por los estudiantes en los diferentes aspectos clave de la enseñanza del Machine Learning, así como en la evaluación de la interacción y participación de los estudiantes en los canales de comunicación y en el acceso y utilización de los recursos educativos proporcionados.
