```python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde el archivo CSV
path = 'C:/Users/marco/Documentos/docencia/groups_list/fundamentos_aritmetica.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float)  # Asegurar que la columna 'Parcial' es float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    global lista_estudiantes
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        # Validación de entrada
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.3
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        elif asistencia != 'n':
            print("Entrada inválida. Por favor, introduce 's' para sí o 'n' para no.")
            return  # Volver a preguntar para el mismo estudiante si la entrada es inválida
        
        # Remover al estudiante de la lista para no llamarlo de nuevo
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Función para guardar cambios al final de la sesión
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()

# Guardar los cambios una vez que todos los estudiantes hayan sido llamados
guardar_cambios()

```

El acrónimo STEAMS incluye:



1. **Science (Ciencia):** Fomenta la curiosidad por el mundo natural y el desarrollo de habilidades científicas, como la observación, experimentación y análisis.



2. **Technology (Tecnología):** Integra herramientas tecnológicas para potenciar el aprendizaje y la resolución de problemas, promoviendo la alfabetización digital desde edades tempranas.

6. **Social Sciences (Ciencias Sociales):** Incorpora la comprensión del entorno social, político, cultural y económico, promoviendo el desarrollo de la conciencia crítica sobre el mundo y su historia, además de habilidades de empatía y ciudadanía.



El enfoque STEAMS pretende desarrollar un aprendizaje **interdisciplinario** que integre estas áreas para fomentar un pensamiento más holístico y práctico.

### **Metodología de Enseñanza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**



El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodología educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en inglés)**. Es especialmente útil para la formación de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades prácticas con el pensamiento crítico y reflexivo. A continuación, se presenta una descripción adaptada a la enseñanza de las matemáticas en educación infantil, Licenciatura en Matemáticas, Ingeniería Agropecuaria y Administración de Empresas:

---



### **1. Fundamentos del ABPP**
El ABPP se basa en:
- **Resolución de Problemas Reales:** Los estudiantes analizan y resuelven problemas auténticos relacionados con su área de formación, fomentando la aplicación práctica del conocimiento.
- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.
- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigación, diseño y solución.

---



### **2. Aplicación del ABPP en Áreas Específicas**

#### **Educación Infantil (Licenciatura en Educación Infantil)**
- **Enfoque:** Desarrollo del pensamiento matemático en la infancia desde contextos lúdicos, familiares y culturales.
- **Problemas típicos:** 
  - ¿Cómo enseñar los números naturales mediante actividades prácticas y juegos?
  - ¿Qué estrategias pueden emplearse para abordar el concepto de cantidad en edades tempranas?
- **Proyectos:** Diseño de secuencias didácticas que usen Python, Jupyter Notebooks y materiales manipulativos para enseñar matemáticas de forma interactiva.
- **Competencias desarrolladas:** 
  - Uso de herramientas tecnológicas para diseñar actividades educativas.
  - Comprensión del desarrollo cognitivo infantil y su relación con las matemáticas.



#### **Licenciatura en Matemáticas**
- **Enfoque:** Resolución de problemas matemáticos aplicados y profundización teórica en áreas como análisis matemático y modelado.
- **Problemas típicos:**
  - ¿Cómo se pueden resolver ecuaciones diferenciales aplicadas a modelos físicos o económicos?
  - ¿Cómo modelar fenómenos naturales mediante funciones matemáticas?
- **Proyectos:** 
  - Desarrollo de software educativo para explorar conceptos matemáticos.
  - Aplicación de modelado matemático para resolver problemas reales en otras disciplinas.
- **Competencias desarrolladas:**
  - Dominio de la matemática teórica y aplicada.
  - Uso de tecnología para la enseñanza y la investigación.



#### **Ingeniería Agropecuaria**
- **Enfoque:** Modelado matemático y análisis de datos para mejorar la producción agrícola y la gestión de recursos.
- **Problemas típicos:**
  - ¿Cómo modelar el crecimiento de cultivos considerando variables climáticas?
  - ¿Cómo optimizar el uso del agua en sistemas de riego mediante análisis matemático?
- **Proyectos:** 
  - Desarrollo de modelos predictivos con Python y bibliotecas de aprendizaje automático.
  - Implementación de proyectos que combinen matemáticas y ciencias de datos para resolver problemas agrícolas.
- **Competencias desarrolladas:**
  - Integración de matemáticas con tecnologías avanzadas.
  - Resolución de problemas agropecuarios mediante análisis y modelado.



#### **Administración de Empresas**
- **Enfoque:** Uso de las matemáticas y la ciencia de datos para la toma de decisiones estratégicas.
- **Problemas típicos:**
  - ¿Cómo identificar patrones de comportamiento del consumidor usando series de tiempo?
  - ¿Qué estrategias matemáticas pueden optimizar la asignación de recursos en una empresa?
- **Proyectos:** 
  - Implementación de modelos de análisis financiero y simulaciones.
  - Desarrollo de dashboards interactivos para la gestión de datos empresariales.
- **Competencias desarrolladas:**
  - Habilidad para interpretar datos matemáticos en contextos de negocio.
  - Capacidad para diseñar soluciones basadas en datos.

---



### **3. Beneficios del ABPP**
- **Desarrollo de competencias clave:**
  - Pensamiento crítico y creativo.
  - Trabajo colaborativo.
  - Comunicación efectiva.
  - Uso de herramientas tecnológicas avanzadas.
- **Adaptación al contexto laboral:** Los estudiantes aprenden a enfrentar desafíos del mundo real y a diseñar soluciones prácticas y aplicables.
- **Flexibilidad e interdisciplinariedad:** La metodología se adapta a diversas áreas, fomentando el aprendizaje integrado.

---



### **4. Ejemplo General de Implementación**
1. **Identificación del problema:** Se plantea un desafío concreto, como modelar el crecimiento poblacional de una comunidad rural (Ingeniería Agropecuaria) o diseñar una estrategia para enseñar fracciones a niños de 8 años (Educación Infantil).
2. **Formación de equipos:** Los estudiantes trabajan en grupos, distribuyendo roles y tareas.
3. **Investigación y diseño:** Los estudiantes investigan el problema, proponen soluciones y diseñan un proyecto.
4. **Implementación:** Se desarrolla el proyecto, como una herramienta tecnológica o una propuesta pedagógica.
5. **Evaluación:** Los estudiantes presentan sus resultados en un informe o una exposición, recibiendo retroalimentación de sus compañeros y docentes.

### Diseño de clase: **Las cuatro ramas del aprendizaje automático**

#### **1. Título de la clase**  
Explorando las cuatro ramas del aprendizaje automático con Python: visualizaciones, animaciones y experimentación en Jupyter Notebooks.  

#### **2. Objetivo de aprendizaje**  
El estudiante será capaz de:  
- Comprender las cuatro ramas principales del aprendizaje automático: aprendizaje supervisado, no supervisado, auto-supervisado y por refuerzo.  
- Implementar ejemplos básicos de cada rama utilizando Python en cuadernos Jupyter, con animaciones y visualizaciones interactivas para experimentar con datos y modelos.  

#### **3. Contenido de la clase**  
1. Introducción a las cuatro ramas del aprendizaje automático:  
   - **Aprendizaje supervisado:** Clasificación y regresión.  
   - **Aprendizaje no supervisado:** Reducción de dimensionalidad y agrupamiento.  
   - **Aprendizaje auto-supervisado:** Autoencoders y predicción basada en secuencias.  
   - **Aprendizaje por refuerzo:** Agentes que toman decisiones en entornos simulados.  

2. Herramientas para la experimentación en Python:  
   - Creación de funciones de usuario para manejar datos y modelos.  
   - Uso de bibliotecas como `matplotlib`, `seaborn`, `plotly`, y `ipywidgets` para visualizaciones y animaciones.  

#### **4. Actividades de aprendizaje**  

**A. Introducción conceptual interactiva**  
- Presentación breve (10 minutos) sobre las cuatro ramas del aprendizaje automático.  
- Discusión en grupos pequeños sobre ejemplos prácticos de cada rama en la vida diaria (clasificación de correos, agrupamiento de clientes, etc.).  

**B. Experimentación práctica con Jupyter Notebooks**  
1. **Aprendizaje supervisado:**  
   - Implementar un modelo de regresión lineal para predecir precios de viviendas.  
   - Crear una animación que muestre cómo cambia la línea de regresión con cada iteración del entrenamiento.  

2. **Aprendizaje no supervisado:**  
   - Usar un conjunto de datos para agrupar puntos en 2D utilizando K-means.  
   - Visualizar el proceso de agrupamiento con animaciones que muestren cómo los centroides convergen.  

3. **Aprendizaje auto-supervisado:**  
   - Implementar un autoencoder simple para comprimir y reconstruir imágenes.  
   - Crear una visualización interactiva que permita comparar imágenes originales y reconstruidas.  

4. **Aprendizaje por refuerzo:**  
   - Simular un agente básico que aprende a moverse en un laberinto.  
   - Mostrar el progreso del agente utilizando una animación paso a paso.
