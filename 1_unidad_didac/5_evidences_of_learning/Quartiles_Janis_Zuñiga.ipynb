#Que son los quartiles? de al menos un ejemplo de utilización.
Los cuartiles son valores que dividen una muestra de datos en cuatro partes iguales , cada una conteniendo el mismo número de observaciones. Se utilizan comúnmente en estadística para describir la distribución de un conjunto de datos y calcular el rango intercuartílico , que es la diferencia entre el tercer cuartil y el primer cuartil.

Por ejemplo, si tenemos una muestra de 20 observaciones, los cuartiles estarían representados por los valores que dividen la muestra en cuatro partes iguales. El primer cuartil (Q1) sería el valor que deja el 25% de los datos por debajo y el 75% por encima . El segundo cuartil (Q2) es la mediana, que deja el 50% de los datos por debajo y el 50% por encima. El tercer cuartil (Q3) es el valor que deja el 75% de los datos por debajo y el 25% por encima.

La utilización más común de los cuartiles es en la construcción de diagramas de caja y bigotes, que son una forma visual de representar los cuartiles y otros estadísticos descriptivos de una muestra. Estos diagramas son útiles para comparar varias muestras o distribuciones y detectar posibles valores atípicos o valores extremos que se encuentran fuera del rango intercuartílico.

##Aprenderse de memoria los códigos para obtención de datos y visualización tabular.
Para obtener los cuartiles y visualizarlos en una tabla, puedes usar una librería en Python como Pandas. Primero, necesitas cargar tus datos en un objeto de tipo DataFrame de Pandas. Luego, puedes utilizar la función describe() y el método quantile() para obtener las estadísticas descriptivas y los cuartiles respectivamente.

Para cargar tus datos en un DataFrame, puedes usar una de las funciones de Pandas, como read_csv() si tus datos están almacenados en un archivo CSV. Aquí un ejemplo con datos hipotéticos:

```python
import pandas as pd

# carga los datos en un objeto DataFrame
caucasia = pd.read_csv('/content/drive/MyDrive/Base de datos/caucasia.csv')
caucasia
```

```python
# muestra las estadísticas descriptivas
print(caucasia.describe())
```

```python
# calcula los cuartiles y los muestra en una tabla
quartiles = caucasia.quantile([0.25, 0.5, 0.75])
print(quartiles)
```
