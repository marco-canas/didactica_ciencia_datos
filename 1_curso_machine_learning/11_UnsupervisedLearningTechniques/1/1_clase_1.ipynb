# Aprendizaje Basado en Proyectos  

### **Metodología de Enseñanza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**

El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodología educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en inglés)**. Es especialmente útil para la formación de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades prácticas con el pensamiento crítico y reflexivo. A continuación, se presenta una descripción adaptada a la enseñanza de las matemáticas en educación infantil, Licenciatura en Matemáticas, Ingeniería Agropecuaria y Administración de Empresas:

---



### **1. Fundamentos del ABPP**
El ABPP se basa en:
- **Resolución de Problemas Reales:** Los estudiantes analizan y resuelven problemas auténticos relacionados con su área de formación, fomentando la aplicación práctica del conocimiento.
- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.
- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigación, diseño y solución.

---



### **2. Aplicación del ABPP en Áreas Específicas**

#### **Educación Infantil (Licenciatura en Educación Infantil)**
- **Enfoque:** Desarrollo del pensamiento matemático en la infancia desde contextos lúdicos, familiares y culturales.
- **Problemas típicos:** 
  - ¿Cómo enseñar los números naturales mediante actividades prácticas y juegos?
  - ¿Qué estrategias pueden emplearse para abordar el concepto de cantidad en edades tempranas?
- **Proyectos:** Diseño de secuencias didácticas que usen Python, Jupyter Notebooks y materiales manipulativos para enseñar matemáticas de forma interactiva.
- **Competencias desarrolladas:** 
  - Uso de herramientas tecnológicas para diseñar actividades educativas.
  - Comprensión del desarrollo cognitivo infantil y su relación con las matemáticas.

#### **Licenciatura en Matemáticas**
- **Enfoque:** Resolución de problemas matemáticos aplicados y profundización teórica en áreas como análisis matemático y modelado.
- **Problemas típicos:**
  - ¿Cómo se pueden resolver ecuaciones diferenciales aplicadas a modelos físicos o económicos?
  - ¿Cómo modelar fenómenos naturales mediante funciones matemáticas?
- **Proyectos:** 
  - Desarrollo de software educativo para explorar conceptos matemáticos.
  - Aplicación de modelado matemático para resolver problemas reales en otras disciplinas.
- **Competencias desarrolladas:**
  - Dominio de la matemática teórica y aplicada.
  - Uso de tecnología para la enseñanza y la investigación.



#### **Ingeniería Agropecuaria**
- **Enfoque:** Modelado matemático y análisis de datos para mejorar la producción agrícola y la gestión de recursos.
- **Problemas típicos:**
  - ¿Cómo modelar el crecimiento de cultivos considerando variables climáticas?
  - ¿Cómo optimizar el uso del agua en sistemas de riego mediante análisis matemático?
- **Proyectos:** 
  - Desarrollo de modelos predictivos con Python y bibliotecas de aprendizaje automático.
  - Implementación de proyectos que combinen matemáticas y ciencias de datos para resolver problemas agrícolas.
- **Competencias desarrolladas:**
  - Integración de matemáticas con tecnologías avanzadas.
  - Resolución de problemas agropecuarios mediante análisis y modelado.

#### **Administración de Empresas**
- **Enfoque:** Uso de las matemáticas y la ciencia de datos para la toma de decisiones estratégicas.
- **Problemas típicos:**
  - ¿Cómo identificar patrones de comportamiento del consumidor usando series de tiempo?
  - ¿Qué estrategias matemáticas pueden optimizar la asignación de recursos en una empresa?
- **Proyectos:** 
  - Implementación de modelos de análisis financiero y simulaciones.
  - Desarrollo de dashboards interactivos para la gestión de datos empresariales.
- **Competencias desarrolladas:**
  - Habilidad para interpretar datos matemáticos en contextos de negocio.
  - Capacidad para diseñar soluciones basadas en datos.

     - Segmentación de clientes.

   | Aspecto                  | Supervisado                           | No supervisado                         |

### **Algoritmos de clustering**  



#### **K-Means**  



1. **Concepto:**  
   Divide los datos en $k$ grupos minimizando la suma de las distancias al cuadrado entre los puntos y el centroide del grupo.  

2. **Pasos principales:**  
   1. Inicializar $k$ centroides aleatoriamente.  
   2. Asignar cada punto al centroide más cercano.  
   3. Recalcular los centroides.  
   4. Repetir hasta convergencia (cuando los centroides ya no cambian significativamente).  

3. **Código Python:**  


````python
from sklearn.datasets import make_blobs # crea un conjunto de datos de 4 clases
from sklearn.cluster import KMeans # algoritmo de agrupamiento k-means
import matplotlib.pyplot as plt # visualización de los datos y sus agrapaciones 

````

### **2. Implementación de funciones de usuario**  

#### **Escribir una función en Python para implementar el algoritmo K-Means:**  
1. **Código Python:** 

````python
 
import numpy as np

def kmeans_custom(X, k, max_iters=100, tol=1e-4):
    # Inicialización aleatoria de centroides
    np.random.seed(0)
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Asignación de puntos al centroide más cercano
        distances = np.linalg.norm(X[:, None] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Recalcular centroides
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        
        # Verificar convergencia
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids

    return labels, centroids
````

### **2. Crear una función para la detección de anomalías basada en densidad**  

#### **Implementación de la detección de anomalías usando modelos de mezcla Gaussiana:**  
- Definir una función para detectar anomalías en un conjunto de datos.  
- Usar el puntaje de densidad del modelo de mezcla Gaussiana para identificar puntos con baja probabilidad.  

#### **Código Python:**  


````python
from sklearn.mixture import GaussianMixture

def detect_anomalies_gmm(X, contamination=0.05):
    gmm = GaussianMixture(n_components=3, random_state=0)
    gmm.fit(X)
    
    # Calcular la densidad logarítmica
    log_density = gmm.score_samples(X)
    threshold = np.percentile(log_density, contamination * 100)
    
    # Identificar anomalías
    anomalies = X[log_density < threshold]
    return anomalies, log_density

# Detectar anomalías en el conjunto de datos de clientes
anomalies, log_density = detect_anomalies_gmm(data_scaled)
print("Número de anomalías detectadas:", len(anomalies))

# Visualización
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c='blue', s=50, label='Normal')
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', s=50, label='Anomalías')
plt.legend()
plt.title("Detección de anomalías basada en densidad")
plt.show()
````
