## El componente histórico y epistemológico


---



### **1. Contexto Histórico del Clustering y Detección de Anomalías**



#### **Clustering**
- **1948:** J.B. MacQueen introduce el concepto formal del clustering, un proceso para agrupar objetos similares en subconjuntos homogéneos.
- **1967:** Se populariza el algoritmo K-Means, desarrollado inicialmente por Hugo Steinhaus y posteriormente por Stuart Lloyd y E.W. Forgy. Su simplicidad y versatilidad lo convirtieron en un pilar en análisis de datos.
- **Década de 1990:** El avance en bases de datos masivas y minería de datos condujo a la creación de algoritmos más eficientes como DBSCAN y OPTICS, que trabajan con estructuras más complejas y ruido.



#### **Detección de Anomalías**


- **Siglo XIX:** Francis Galton y Karl Pearson desarrollan herramientas estadísticas para estudiar distribuciones. Estas ideas fundamentaron los enfoques iniciales de detección de valores atípicos.
- **1960-1980:** La estadística robusta se establece formalmente para manejar valores atípicos, con John Tukey proponiendo técnicas gráficas como los diagramas de caja (boxplots).
- **Década de 2000:** Con el auge de la inteligencia artificial, se integran métodos avanzados de ML como Isolation Forest, que combina estadísticas con teoría de grafos y enfoques probabilísticos.

---



### **2. Elementos Epistemológicos**



#### **Clustering como Herramienta Epistémica**
- El clustering se basa en la premisa de **identificar patrones en datos no etiquetados**, lo que revela estructuras subyacentes. Este enfoque refleja el método inductivo, utilizado para construir hipótesis a partir de observaciones.
- Ejemplo: En ciencias naturales, el clustering ayuda a clasificar especies o analizar conglomerados galácticos.



#### **Detección de Anomalías: Filosofía de los Errores**
- La detección de anomalías se fundamenta en identificar desviaciones del comportamiento esperado, lo cual está ligado al concepto epistemológico de **error y validación científica**.
- En astronomía, la detección de valores extremos permitió identificar eventos significativos, como explosiones de supernovas, que inicialmente parecían anomalías.

---

- **1990s-2000s:** El desarrollo de bibliotecas de visualización y computación (Matplotlib, Pandas) permitió democratizar herramientas estadísticas y de ML, facilitando experimentos en clustering y anomalías.

### **Algoritmos de clustering**  



#### **K-Means**  



1. **Concepto:**  
   Divide los datos en $k$ grupos minimizando la suma de las distancias al cuadrado entre los puntos y el centroide del grupo.  



2. **Pasos principales:**  
   1. Inicializar $k$ centroides aleatoriamente.  
   2. Asignar cada punto al centroide más cercano.  
   3. Recalcular los centroides.  
   4. Repetir hasta convergencia (cuando los centroides ya no cambian significativamente).  

3. **Código Python:**

### **DBSCAN**  

1. **Concepto:**  
   Agrupa puntos densamente conectados basándose en dos parámetros:  
   - **$\varepsilon$:** Radio para buscar vecinos.  
   - **min_samples:** Número mínimo de puntos necesarios para formar un grupo.  



2. **Ventajas:**  
   - Detecta clusters de forma no esférica.  
   - Identifica puntos como ruido.  

3. **Código Python:**  


````python
from sklearn.cluster import DBSCAN
import numpy as np

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualización
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', s=50)
plt.title("DBSCAN Clustering")
plt.savefig('C:/Users/marco/Documentos/investigacion/didactica_ciencia_datos/1_curso_machine_learning/11_UnsupervisedLearningTechniques/1/images/dbscsn.jpg')
plt.grid(alpha = 0.3)
plt.show()
 
````

## **Visualización y animación**  



1. **Uso de `matplotlib` y `plotly` para graficar clusters y anomalías:**  
   - `matplotlib`: Para gráficos básicos.  
   - `plotly`: Para visualizaciones interactivas.  



2. **Animación con `matplotlib.animation`:**  
   - Ejemplo de animación del proceso iterativo de K-Means:  


````python

import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Crear datos sintéticos
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Configuración inicial de la figura
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(X[:, 0].min() - 1, X[:, 0].max() + 1)
ax.set_ylim(X[:, 1].min() - 1, X[:, 1].max() + 1)
scatter = ax.scatter([], [], s=50, cmap='viridis', alpha=0.6)
centroids = ax.scatter([], [], s=200, c='red', edgecolor='black', alpha=0.75)

# Inicializar modelo de KMeans
kmeans = KMeans(n_clusters=4, init='random', n_init=1, random_state=0)

# Función de actualización para la animación
def update(frame):
    # Ajustar el modelo de KMeans para iteraciones adicionales
    kmeans.max_iter = frame + 1
    kmeans.fit(X)
    
    # Actualizar el gráfico con los nuevos datos
    scatter.set_offsets(X)
    scatter.set_array(kmeans.labels_)
    centroids.set_offsets(kmeans.cluster_centers_)
    ax.set_title(f"Iteración {frame + 1}", fontsize=14)

# Crear animación
ani = FuncAnimation(fig, update, frames=10, repeat=False)

# Mostrar la animación en Jupyter Notebook
from IPython.display import HTML
HTML(ani.to_jshtml())
````

### **2. Implementación de funciones de usuario**  



#### **Escribir una función en Python para implementar el algoritmo K-Means:**  
1. **Código Python:** 

````python
 
import numpy as np

def kmeans_custom(X, k, max_iters=100, tol=1e-4):
    # Inicialización aleatoria de centroides
    np.random.seed(0)
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Asignación de puntos al centroide más cercano
        distances = np.linalg.norm(X[:, None] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Recalcular centroides
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        
        # Verificar convergencia
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids

    return labels, centroids
````


## **3. Experimentación con algoritmos**  

#### **Aplicar K-Means, DBSCAN y modelos de mezcla Gaussiana al conjunto de datos Iris:**  
1. **Código Python:**  


````python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA

# Reducir dimensiones para visualización
pca = PCA(n_components=2)
X_pca = pca.fit_transform(iris.data)

# K-Means
kmeans = KMeans(n_clusters=3, random_state=0).fit(iris.data)
kmeans_labels = kmeans.labels_

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5).fit(iris.data)
dbscan_labels = dbscan.labels_

# Modelos de mezcla Gaussiana
gmm = GaussianMixture(n_components=3, random_state=0).fit(iris.data)
gmm_labels = gmm.predict(iris.data)

# Visualización en 2D
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
algorithms = [('K-Means', kmeans_labels), ('DBSCAN', dbscan_labels), ('GMM', gmm_labels)]

for ax, (name, labels) in zip(axes, algorithms):
    ax.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
    ax.set_title(name)
plt.suptitle("Resultados de clustering en el conjunto de datos Iris")
plt.show()
 
````

### **4. Visualización y animación**  

#### **Crear una animación que muestre el proceso iterativo del algoritmo K-Means:**  
1. **Código Python:**  


````python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

def update(frame):
    labels, centroids = kmeans_custom(iris.data, k=3, max_iters=frame + 1)
    ax.clear()
    ax.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
    ax.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75)

ani = FuncAnimation(fig, update, frames=10, repeat=False)
plt.show()
 
````

### **2. Crear una función para la detección de anomalías basada en densidad**  

#### **Implementación de la detección de anomalías usando modelos de mezcla Gaussiana:**  
- Definir una función para detectar anomalías en un conjunto de datos.  
- Usar el puntaje de densidad del modelo de mezcla Gaussiana para identificar puntos con baja probabilidad.  

#### **Código Python:**  


````python
from sklearn.mixture import GaussianMixture

def detect_anomalies_gmm(X, contamination=0.05):
    gmm = GaussianMixture(n_components=3, random_state=0)
    gmm.fit(X)
    
    # Calcular la densidad logarítmica
    log_density = gmm.score_samples(X)
    threshold = np.percentile(log_density, contamination * 100)
    
    # Identificar anomalías
    anomalies = X[log_density < threshold]
    return anomalies, log_density

# Detectar anomalías en el conjunto de datos de clientes
anomalies, log_density = detect_anomalies_gmm(data_scaled)
print("Número de anomalías detectadas:", len(anomalies))

# Visualización
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c='blue', s=50, label='Normal')
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', s=50, label='Anomalías')
plt.legend()
plt.title("Detección de anomalías basada en densidad")
plt.show()
````
