```python
import pandas as pd 
```

```python
df_a = pd.DataFrame({'Producto':['A', 'B', 'C'], 'Ventas':[10, 20, 30]})
df_a
```

```python
df_b = pd.DataFrame({'Producto':['D', 'E', 'F'], 'Ventas': [40, 50, 60]})
df_b
```

```python
df_concat = pd.concat([df_a, df_b], axis = 0, join = 'outer', ignore_index = True) 
df_concat 
```
