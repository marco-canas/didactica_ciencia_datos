# Putting it all together.

Let's apply everything we learned and try to predict the S&P 500.

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
df = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/SPY?period1=728265600&period2=1702166400&interval=1d&events=history&includeAdjustedClose=true")

# Set the date as the index
df['date'] = pd.to_datetime(df['Date'])
df.set_index('date', inplace=True)
df = df['Close']

# Calculate monthly average temperature
Stock_Price = df.resample('D').mean()

# Display the first few rows of the monthly average temperature
Stock_Price.head()
```
