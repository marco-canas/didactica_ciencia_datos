# Constitución de un clasificador binario utilizando la clase Perceptron de Sklearn

```python
import numpy as np 
from matplotlib import pyplot as plt 

from sklearn.datasets import load_iris

from sklearn.linear_model import Perceptron 



```

# Obtener los datos 

```python
iris = load_iris() 
```

```python
iris.keys() 
```

```python
iris.filename 
```

```python
X = iris.data
```

```python
X.shape 
```

```python
X = X[:, (2,3)]
```

```python
y = (iris.target == 0) 
```

```python
 y = y.astype(np.int64)  
```

```python
X[:,0] 
```

```python
type(y[0])
```

# Graficación de los datos 

```python
fig, ax  = plt.subplots()

ax.set(title = 'Datos de florez de iris', xlabel = 'longitud del pétalo', ylabel = 'ancho de pétalo')

p, q, r, s = [ np.min(X[:,0])-0.4, np.max(X[:,0]) + 0.4, np.min(X[:,1]) - 0.4, np.max(X[:,1]) + 0.4 ] 
plt.axis([p,q,r,s]) 

plt.plot(X[:,0][y ==  1], X[:,1][y ==  1], 'ro') 

plt.plot(X[:,0][y ==  0], X[:,1][y ==  0], 'go') 

plt.grid()
```

# Determinemos nuestro clasificador binario 

```python
per_clr = Perceptron() 
```

```python
per_clr.fit(X,y)
```

```python
w0, w1, w2 = per_clr.intercept_[0], per_clr.coef_[0][0], per_clr.coef_[0][1]
```

```python
b, m = -w0/w2, -w1/w2 
```

```python
fig, ax  = plt.subplots()

ax.set(title = 'Datos de florez de iris', xlabel = 'longitud del pétalo', ylabel = 'ancho de pétalo')

p, q, r, s = [ np.min(X[:,0])-0.4, np.max(X[:,0]) + 0.4, np.min(X[:,1]) - 0.4, np.max(X[:,1]) + 0.4 ] 
plt.axis([p,q,r,s]) 

x0 = [p,q]
threshold = [p*m+b, q*m+b]

plt.plot(x0, threshold, 'k-', label = 'Clasificador') 

plt.plot(X[:,0][y ==  1], X[:,1][y ==  1], 'go', label = 'Setosa') 

plt.plot(X[:,0][y ==  0], X[:,1][y ==  0], 'ro', label = 'no Setosa') 

plt.grid()

plt.legend()

plt.savefig('clasificacion_con_perceptron.jpg')

plt.show() 


```
