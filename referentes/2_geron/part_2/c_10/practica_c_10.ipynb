# Constitución de un clasificador binario utilizando la clase Perceptron de Sklearn

```python
import numpy as np 
from sklearn.datasets import load_iris 
```

```python
iris = load_iris() 
```

```python
type(iris) 
```

```python
iris.keys()
```

```python
X = iris.data 
```

```python
y = iris.target 
```

```python
iris.target_names
```

```python
y 
```

```python
X=X[:, (2,3)]
X 
```

```python
X_y = np.c_[X,y] 
```

```python
X_y
```

```python
import pandas as pd 
```

```python
tabla_bonita = pd.DataFrame(X_y, columns = ['longitud del pétalo', 'ancho del pétalo', 'Tipo de flor'], index = np.arange(1, 151)) 
```

```python
tabla_bonita
```
