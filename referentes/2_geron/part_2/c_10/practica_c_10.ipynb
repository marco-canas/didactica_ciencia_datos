# Clasificador binario: Detector de setosa 

```python
import numpy as np 
from sklearn.datasets import load_iris 
```

```python
iris = load_iris() 
```

```python
type(iris)
```

```python
iris.data.shape  # shape significa forma 
```

```python
iris.feature_names
```

```python
iris.target 
```

```python
iris.target_names
```

```python
X = iris.data[:, (2,3)]
y = iris.target 
```

# Ahora vamos a visualizar los datos 

```python
import matplotlib.pyplot as plt 

plt.figure() 

plt.title("Visualizar a las 150 flores de iris")  # Lo haremos es un gráfico de dispersión 

plt.xlabel("Longitud del pétalo (cm)")

plt.ylabel('Ancho del pétalo (cm)') 

plt.scatter(X[:,0][y==0], X[:,1][y==0], color = 'purple', label = 'Setosa') 

plt.scatter(X[:,0][y==1], X[:,1][y==1], color = 'red', label = 'No setosa')  
plt.scatter(X[:,0][y==2], X[:,1][y==2], color = 'red') 

plt.legend()

plt.show() 
```

## Inferencias o interpretaciones del gráfico

* La setosa es un tipo de flor de iris de tamaño de pétalo pequeño. 

# Ahora si vamos a clasificar

```python
from sklearn.linear_model import Perceptron
```

```python
clasificador = Perceptron() # instanciar la clase Perceptron 

clasificador.fit(X, y) # con esta línea el ha aprendido a clasificar. 
```

```python
clasificador.intercept_[0]
```

```python
w0, w1, w2 = clasificador.intercept_[0], clasificador.coef_[0][0], clasificador.coef_[0][1] 
```

```python
pendiente, intersecto = -w1/w2, -w0/w2
```

```python
pendiente 
```

```python
intersecto 
```

```python
import matplotlib.pyplot as plt 

plt.figure() 

plt.title("Visualizar a las 150 flores de iris")  # Lo haremos es un gráfico de dispersión 

plt.xlabel("Longitud del pétalo (cm)")

plt.ylabel('Ancho del pétalo (cm)') 

plt.scatter(X[:,0][y==0], X[:,1][y==0], color = 'purple', label = 'Setosa') 

plt.scatter(X[:,0][y==1], X[:,1][y==1], color = 'red', label = 'No setosa')  
plt.scatter(X[:,0][y==2], X[:,1][y==2], color = 'red') 

minimo, maximo = min(X[:,0]), max( X[:,0]) 

plt.plot([minimo, maximo], [minimo*pendiente + intersecto, maximo*pendiente+ intersecto])

plt.legend()

plt.savefig('clasificador_binario_setosa.jpg')

plt.show() 
```
