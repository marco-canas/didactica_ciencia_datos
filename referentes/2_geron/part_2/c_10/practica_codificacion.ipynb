```python
from keras.datasets import fashion_mnist

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
def save_fig(fig_id, tight_layout = True, fig_extension = 'png', resolution = 300):
    path = os.path.join(camino_imagen, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
class_names = ['Camiseta', 'Pantalón', 'Jersey', 'Vestido', 'Abrigo',  \
    'Sandalia', 'Camisa', 'Zapatilla de deporte', 'Bolso', 'Botín']
```

```python
plt.figure(figsize=(n_columnas*10, n_filas*10))

for fila in range(n_filas):
    for columna in range(n_columnas): 
        indice = n_columnas*fila + columna
        plt.subplot(n_filas, n_columnas, indice + 1)
        plt.imshow(x_train[indice], cmap = 'binary', interpolation='nearest')
        plt.axis('off') 
        plt.title(class_names[y_train[indice]], fontsize = 12)
plt.subplots_adjust(wspace=0.2, hspace=0.5)         
save_fig('Todas_las_imagenes_que joseph_pidio', tight_layout = False)
plt.show() 
```

```python
from tensorflow import keras 
```

```python
model = keras.models.Sequential() 

model.add(keras.layers.Flatten(input_shape = [28, 28]))
model.add(keras.layers.Dense(300, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```
