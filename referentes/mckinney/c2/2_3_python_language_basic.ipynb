### Ejemplo de for loop 

```python
import numpy as np 
x = 3
less = []
greater = []
array = np.array([1,2,3,4,5,6])
pivot = 3

for x in array:
    if x < pivot:
        less.append(x)
    else:
        greater.append(x)     
```
